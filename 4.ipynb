# Шаг 2. Импортируйте набор данных с этого [адреса]  (https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

```python
download= ('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
```

# Шаг 3. Сохраните его в переменной с именем drinks

```python
drinks= pd.read_csv(download)
pd.set_option("display.max.columns", None)
drinks.head(10)
```

# Шаг 4. На каком континенте в среднем пьют больше пива?

```python
most_beer_drinks= drinks.sort_values(by='beer_servings', ascending= False)
print(most_beer_drinks['continent'][1])
```

# Шаг 5.  Для каждого континента вывести статистику по потреблению вина.

```python
most_vine_drinks= drinks.drop(drinks.columns[[1, 2, 4]], axis=1) 
most_vine_drinks.drop_duplicates(subset=['continent'], keep='first')
most_vine_drinks_AS= most_vine_drinks[(most_vine_drinks['continent']=='AS')]
most_vine_drinks_EU= most_vine_drinks[(most_vine_drinks['continent']=='EU')]
most_vine_drinks_AF= most_vine_drinks[(most_vine_drinks['continent']=='AF')]
most_vine_drinks_SA= most_vine_drinks[(most_vine_drinks['continent']=='SA')]
most_vine_drinks_OC= most_vine_drinks[(most_vine_drinks['continent']=='OC')]
most_vine_drinks_NaN= most_vine_drinks[most_vine_drinks['continent'].isnull()]
most_vine_drinks_AS,most_vine_drinks_EU,most_vine_drinks_AF,most_vine_drinks_SA,most_vine_drinks_OC,most_vine_drinks_NaN
```

# Шаг 6. Вывести среднее потребление алкоголя по каждому континенту по каждому столбцу

```python
drinks_mean= drinks.drop(drinks.columns[[0 , 4]], axis=1) 
drinks_mean
drinks_mean_AS= drinks_mean[(drinks_mean['continent']=='AS')].describe().loc['mean']
drinks_mean_EU= drinks_mean[(drinks_mean['continent']=='EU')].describe().loc['mean']
drinks_mean_AF= drinks_mean[(drinks_mean['continent']=='AF')].describe().loc['mean']
drinks_mean_SA= drinks_mean[(drinks_mean['continent']=='SA')].describe().loc['mean']
drinks_mean_OC= drinks_mean[(drinks_mean['continent']=='OC')].describe().loc['mean']
drinks_mean_NaN= drinks_mean[drinks_mean['continent'].isnull()].describe().loc['mean']
print('AS:\n',drinks_mean_AS ,'\n' ,'EU:\n',drinks_mean_EU,'\n', 'AF:', drinks_mean_AF,'\n','SA:\n', drinks_mean_SA,'\n','OC:\n', drinks_mean_OC,'\n','NaN:\n', drinks_mean_NaN)
```

# Шаг 7.  Вывести медианное потребление алкоголя по каждому континенту по каждому столбцу

```python
drinks_median_AS= drinks_mean[(drinks_mean['continent']=='AS')].describe().loc['50%']
drinks_median_EU= drinks_mean[(drinks_mean['continent']=='EU')].describe().loc['50%']
drinks_median_AF= drinks_mean[(drinks_mean['continent']=='AF')].describe().loc['50%']
drinks_median_SA= drinks_mean[(drinks_mean['continent']=='SA')].describe().loc['50%']
drinks_median_OC= drinks_mean[(drinks_mean['continent']=='OC')].describe().loc['50%']
drinks_median_NaN= drinks_mean[drinks_mean['continent'].isnull()].describe().loc['50%']
print('AS:\n',drinks_median_AS ,'\n' ,'EU:\n',drinks_median_EU,'\n', 'AF:', drinks_median_AF,'\n','SA:\n', drinks_median_SA,'\n','OC:\n', drinks_median_OC,'\n','NaN:\n', drinks_median_NaN)
```

# Шаг 8.  Вывести среднее, минимальное и максимальное значения для потребления напитков(столбец spirit).

```python
drinks_spirit= drinks.drop(drinks.columns[[0, 1, 3, 4]], axis=1)
drinks_spirit_AS= drinks_spirit[(drinks_spirit['continent']=='AS')].describe().loc['mean':'max'].drop(labels = ['std','25%','50%','75%'],axis = 0)
drinks_spirit_EU= drinks_spirit[(drinks_spirit['continent']=='EU')].describe().loc['mean':'max'].drop(labels = ['std','25%','50%','75%'],axis = 0)
drinks_spirit_AF= drinks_spirit[(drinks_spirit['continent']=='AF')].describe().loc['mean':'max'].drop(labels = ['std','25%','50%','75%'],axis = 0)
drinks_spirit_SA= drinks_spirit[(drinks_spirit['continent']=='SA')].describe().loc['mean':'max'].drop(labels = ['std','25%','50%','75%'],axis = 0)
drinks_spirit_OC= drinks_spirit[(drinks_spirit['continent']=='OC')].describe().loc['mean':'max'].drop(labels = ['std','25%','50%','75%'],axis = 0)
drinks_spirit_NaN= drinks_spirit[drinks_spirit['continent'].isnull()].describe().loc['mean':'max'].drop(labels = ['std','25%','50%','75%'],axis = 0)
print('AS:\n',drinks_spirit_AS ,'\n' ,'EU:\n',drinks_spirit_EU,'\n', 'AF:', drinks_spirit_AF,'\n','SA:\n', drinks_spirit_SA,'\n','OC:\n', drinks_spirit_OC,'\n','NaN:\n', drinks_spirit_NaN)
```
